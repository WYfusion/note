### 有序的字典 $collections$库中的$OrderedDict()$方法

```python
import collections
import torch.nn as nn
```

```python
dic = collections.OrderedDict()
dic['conv1'] = nn.Conv2d(1, 6, 5)
dic['pool1'] = nn.MaxPool2d(2, 2)
dic['conv2'] = nn.Conv2d(6, 16, 5)
dic['pool2'] = nn.MaxPool2d(2, 2)
dic['fc1'] = nn.Linear(16 * 5 * 5, 120)
dic['fc2'] = nn.Linear(120, 84)
dic['fc3'] = nn.Linear(84, 10)
dic['softmax'] = nn.Softmax(dim=1)

print(dic)
```

##### 使用.items()方法遍历OrderedDict

```python
for key, value in dic.items():
    print(key, value)
```

### 构建神经网络

```python
features = nn.Sequential(collections.OrderedDict([
            ('conv', nn.Conv2d(3, 32, kernel_size=7, stride=2, padding=3, bias=False)),
            ('norm', nn.BatchNorm2d(32)),
            ('relu', nn.ReLU(inplace=True)),
            ('pool', nn.MaxPool2d(kernel_size=3, stride=2, padding=1)),
            ]))

print(features)
```

##### 常规构建神经网络的方法

```python
features = nn.Sequential(
    nn.Conv2d(3, 32, kernel_size=7, stride=2, padding=3, bias=False),
    nn.BatchNorm2d(32),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
)
print(features)
```
