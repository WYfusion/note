### 导入必要的软件包

```python
import torch
from torchvision.models import AlexNet  # 导入AlexNet官方模型
import os
```

#### 设置路径

```python
path_root = os.path.join(os.getcwd(), 'trained_models')
if not os.path.exists(path_root):
    os.makedirs(path_root, exist_ok=True)
```

##### 保存整个模型

```python
path = os.path.join(path_root, 'alexnet_model_full.pth')
model = AlexNet().to('cuda'if torch.cuda.is_available() else 'cpu')
torch.save(model, path)
```

##### 只保存模型训练的参数（不包括网络结构）

```python
path = os.path.join(path_root,'alexnet_model_params.pth')
torch.save(model.state_dict(), path)
```

### load 方法的map_location参数的使用
使用 map_location 参数对 GPU 和 CPU 进行转化

1）GPU 转 CPU

```python
model = torch.load(path, map_location='cpu')
```

2）CPU 转 GPU

```python
model = torch.load(path, map_location=lambda storage, loc: storage.cuda(0))
```

3）GPU之间转换

```python
torch.load(path, map_location={'cuda:0':'cuda:1'})    # GPU0转到GPU1
```
